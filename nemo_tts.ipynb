```python
# Import libraries
import torch
import soundfile as sf
from nemo.collections.tts.models import HifiGanModel, Tacotron2Model

# Load pre-trained Tacotron2 and HiFi-GAN models
tts_model = Tacotron2Model.from_pretrained(model_name="tts_en_tacotron2")
vocoder = HifiGanModel.from_pretrained(model_name="tts_en_hifigan")
```

```python
def text_to_speech(text, output_file="output.wav"):
    """
    Converts input text to speech and saves the audio as a .wav file.

    Args:
        text (str): The input text to convert to speech.
        output_file (str): The name of the output audio file (default: "output.wav").
    """
    # Ensure models are in evaluation mode
    tts_model.eval()
    vocoder.eval()

    # Generate mel-spectrogram from text
    with torch.no_grad():
        # Parse text to tokens
        tokens = tts_model.parse(text=text)

        # Generate mel-spectrogram
        mel_output = tts_model.generate_spectrogram(tokens=tokens)

        # Convert mel-spectrogram to audio
        # Pass mel_output as the 'spec' argument
        audio = vocoder.convert_spectrogram_to_audio(spec=mel_output)

    # Save the generated audio to a file
    sf.write(output_file, audio[0].cpu().numpy(), 22050)  # 22,050 Hz sampling rate
    print(f"Audio saved as {output_file}")
```

```python
text_input = "Hello, I am Praveen working as an AI intern at Atmecs Global."


text_to_speech(text_input)
```
